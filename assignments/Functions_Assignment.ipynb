### Q3. Write a Python function to **multiply all the numbers in a list**.

Sample List: `(8, 2, 3, -1, 7)` → Expected Output: `-336`

````python
# Write your code here
def mul(l):
    result = 1
    for n in l:
        result *= n
    return result
l = [8, 2, 3, -1, 7]
mul(l)
````

### Q6. Write a Python function that accepts a string and **counts the number of upper and lower case letters**.

Sample: `'The quick Brow Fox'` → Upper: 3, Lower: 12

````python
# Write your code here
def count_letter(l):
    Upper = []
    lower = []
    for i in l:
        if i.isupper():
            Upper.append(i)
        elif i.islower():
            lower.append(i)
    print(f"Upper: {len(Upper)}, Lower: {len(lower)}")

count_letter('The quick Brow Fox')
````

### Q7. Write a Python function that takes a list and returns a new list with **distinct elements**.

Sample: `[1,2,3,3,3,3,4,5]` → `[1, 2, 3, 4, 5]`

````python
# Write your code here
def is_distinct(l):
    return set(l)
l = [1,2,3,3,3,3,4,5]
is_distinct(l)
````

### Q8. Write a Python function that checks whether a passed string is a **palindrome** or not.

````python
# Write your code here
def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome('madam'))
print(is_palindrome('mad'))
````

### Q10. Write a Python program to detect the **number of local variables** declared in a function.

````python
# Write your code here
def count_local_variables(func):
    return func.__code__.co_nlocals
def example_func(a, b):
    c = a + b
    d = c * 2
    return d

print(count_local_variables(example_func))  # Output: 4
````
