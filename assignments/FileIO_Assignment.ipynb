### Q5. Write a Python program to **read a file line by line and store it into a list**.

### Q7. Write a Python program to **find the longest words** in a file.

```python
# Write your code here
with open("Sanket.txt",'r')as file:
    words = file.read().split()
if not words:
    print("File is empty")
else:
    long_word = max(words, key=len)


print(f"Longest Word: {long_word}")
print(f"length : {len(long_word)}")
```

### Q8. Write a Python program to **count the number of lines** in a text file.

```python
# Write your code here
with open("Sanket.txt",'r')as file:
    line = sum(1 for line in file)
print(f"Total lines : {line}")
```

### Q9. Write a Python program to **count the frequency of words** in a file.

```python
# Write your code here
word_freq = {}
with open("Sanket.txt",'r')as file:
    for line in file:
        words = line.lower().split()
        for word in words:
            word_freq[word]=word_freq.get(word,0)+1
print(word_freq)
```

### Q13. Write a Python program to **read a random line** from a file.

```python
# Write your code here
import random
with open('Sanket.txt') as f:
    lines = f.readlines()
    print(random.choice(lines).strip())
```
