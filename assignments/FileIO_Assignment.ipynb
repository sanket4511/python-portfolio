### Q5. Write a Python program to **read a file line by line and store it into a list**.

```python
# Write your code here
with open("Sanket.txt","r") as f:
    content_list = f.readlines()
print(content_list)
content_list = [x.strip() for x in content_list]
print(content_list)
```
