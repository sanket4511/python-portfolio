## Question 1

Write a Python program to handle a ZeroDivisionError exception when dividing a number by zero.

```python
# Write your code here
a =  int(input("Enter num1"))
b =  int(input("Enter an num2"))

try:
    res = a/b
except ZeroDivisionError:
    res = 0
print(res)
```

## Question 3

Write a Python program that opens a file and handles a FileNotFoundError exception if the file does not exist.

```python
# Write your code here
try:
    with open("sanket.txt",'r')as file:
        content = file.read()
except FileNotFoundError:
    print("The specified file was not found")
```

## Question 4

Write a Python program that prompts the user to input two numbers and raises a TypeError exception if the inputs are not numerical.

```python
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    print("Sum:", num1 + num2)

except ValueError:
    raise TypeError("Inputs must be numerical values only.")
```

## Question 5

Write a Python program that opens a file and handles a PermissionError exception if there is a permission issue.

```python
# Write your code here
try:
    with open("sample.txt",'r')as file:
        content = file.read()
        print(content)
except PermissionError:
    print("permission Denied")
except FileNotFoundError:
    print("file not found")
```

## Question 6

Write a Python program that executes an operation on a list and handles an IndexError exception if the index is out of range.

```python
# Write your code here
try:
    numbers = [2,3,4,5,6,10]
    index = int(input("Enter an index: "))

    result  = numbers[index]**2 
    print("Square of element:",result)

except IndexError:
    print("Index error, Index Out of Range")
except ValueError:
    print("plz enter a valid number")
```
