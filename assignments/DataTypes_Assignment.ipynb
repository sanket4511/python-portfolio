### Q5. For each element, compute the **sum of all elements greater than or equal to it**.

Input: `[2,4,6,10,1]` → Output: `[22,20,16,10,23]`

```python
# Write your code here
def sum_greater_or_equal(arr):
    result = []
    for element in arr:
        
        total = sum(x for x in arr if x >= element)
        result.append(total)
    return result
input_arr = [2, 4, 6, 10, 1]
output = sum_greater_or_equal(input_arr)
print(f"Input: {input_arr}")
print(f"Output: {output}")
```

### Q16. Write a Python function to check whether a number is **perfect or not**.

A perfect number equals half the sum of all its positive divisors. Examples: 6, 28, 496, 8128.

```python
# Write your code here
def perfect_concise(n):
    return n == sum(i for i in range(1, n) if n % i == 0)

print(perfect_concise(6))  
```

### Q18. Write a Python function that prints a **histogram with bin size 10**.

Input: `[13,42,15,37,22,39,41,50]` → `{11-20:2, 21-30:1, 31-40:2, 41-50:3}`

```python
# Write your code here
def histogram(data):
    hsit ={}
    for num in data:
        lower = ((num -1)//10)*10+1
        upper = lower + 9

        key = f"{lower}-{upper}"

        if key in hsit:
            hsit[key] += 1
        else:
            hsit[key] = 1
    return hsit
data = [13,42,15,37,22,39,41,50]
print(histogram(data))
```
